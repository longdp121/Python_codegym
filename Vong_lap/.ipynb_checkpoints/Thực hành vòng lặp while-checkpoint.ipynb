```python
'''
Khởi động 1: Viết chương trình in ra tất cả các số chia hết cho cả 5 và 7 trong khoảng từ 1500 đến 2700
'''
max_num = 2700
min_num = 1500
while min_num <= max_num:
    if min_num % 5 == 0 and min_num % 7 == 0:
        print(min_num, end = " ")
    min_num += 1
```

```python
'''
Bài 1: Viết chương trình nhập vào chiều cao tam giác vuông và in ra tam giác có hình dạng như sau:
*****
****
***
** 
*
'''
h = int(input())
txt = "*"
i = 0
while i < h:
    shape = txt * (h - i)
    print(shape)
    i += 1
```

```python
'''
Bài 2: Viết chương trình nhập vào chiều cao tam giác vuông và in ra tam giác có hình dạng như sau:
* * * * * *
* * * * *
* * * *
* * *
* *
*
'''
h = int(input())
txt = "*"
i = 0
while i < h:
    shape = (txt + " ") * (h - i)
    print(shape.strip())
    #print(len(shape))
    i += 1
```

```python
'''
Bài 4: Viết chương trình nhập vào một số nguyên n và in ra các số từ 1 đến n với các trường hợp ngoại lệ như sau:
- Nếu số đó chia hết cho 3 thì in "py"
- Nếu số đó chia hết cho 7 thì in "thon"
- Nếu số đó chia hết cho cả 3 và 7 thì in "python"

Ví dụ: n = 25
1
2
py
4
5
py
thon
8
py
10
11
py
13
thon
py
16
17
py
19
20
python
22
23
py
25
'''
num = int(input("Enter any number: "))
i = 1
while i < num + 1:
    if i % 3 == 0 and i % 7 == 0:
        display_num = "python"
    elif i % 7 == 0:
        display_num = "thon"
    elif i % 3 == 0:
        display_num = "py"
    else:
        display_num = i
    i += 1
    print(display_num, end = " ")
```

```python
#### '''
from datetime import datetime
start_time = datetime.now()

num = int(input())
#num = 6700417
#num = 2147483647
if num == 1 or num == 2:
    print(f"{num} la so nguyen to")
else:
    i = 1
    while i < (num - 1):
        i += 1
        check = num % i
        if check == 0:
            num_type = "khong phai so nguyen to"
            break
        else:
            num_type = "la so nguyen to"
    print(f"{num} {num_type}")

end_time = datetime.now()
print('Duration: {}'.format(end_time - start_time))
```

```python
## 
'''
Bài 6: Viết chương trình mô phỏng trò chơi kéo búa bao theo luật chơi như sau:
Quy ước: 1 - búa, 2 - kéo, 3 - bao
Khi trò chơi bắt đầu, yêu cầu người chơi nhập 1 số từ bàn phím đại diện cho kéo, búa, bao
Chương trình tự random ra một lựa chọn từ 1 đến 3.
Nếu người chơi thua thì tiếp tục chơi cho đến khi thắng thì kết thúc chương trình.
Ví dụ:

Mời lựa chọn(1 - búa, 2 - kéo, 3 - bao): 1
Máy tính chọn: 3
Người chơi thua.

Mời lựa chọn(1 - búa, 2 - kéo, 3 - bao): 2
Máy tính chọn: 1
Người chơi thua.

Mời lựa chọn(1 - búa, 2 - kéo, 3 - bao): 1
Máy tính chọn: 2
Người chơi thắng.
'''

from random import randint

result = None

while result != "Thang":
    #Khai bao bien player
    player = str(input())
    if player != "1" and player != "2" and player != "3":
        print("Nhap sai")
        continue
    elif player == "1" or player == "2" or player == "3":
        if player == "1":
            player_choose = "Dam"
        elif player == "2":
            player_choose = "La"
        elif player == "3":
            player_choose = "Keo"
            
        #Khai bao bien com
        com = randint(1,3)
        if com == 1:
            com_choose = "Dam"
        elif com == 2:
            com_choose = "La"
        elif com == 3:
            com_choose = "Keo"

        #Khai bao result
        if player_choose == com_choose:
            result = "Hoa"

        elif player_choose == "Dam":  #Player chon Dam__1
            if com_choose == "La":
                result = "Thua"
            else:
                result = "Thang"

        elif player_choose == "La":  #Player chon La__2
            if com_choose == "Keo":
                result = "Thua"
            else:
                result = "Thang"

        else:  #Player chon Dam__3
            if com_choose == "Dam":
                result = "Thua"
            else:
                result = "Thang"
        print(f"Nguoi choi chon {player_choose}, may tinh chon {com_choose}, ket qua: {result}")
print("GAME OVER")
```
