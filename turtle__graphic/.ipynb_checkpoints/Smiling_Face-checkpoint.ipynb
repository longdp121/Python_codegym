```python
import turtle
pen = turtle.Turtle()
pen.pensize(3)
#khai bao bien
fsize = 15 #face size
msize = fsize*2/3 #mouth size
tsize = fsize/4 #tooth size
esize = tsize/2 #eye size
x = -170
y = -200
y_center = y + fsize
y_mouth = y + fsize*3/4
#face
pen.penup()
pen.goto(x, y) 
pen.pendown()
pen.fillcolor("#F9FD00")
pen.begin_fill()
pen.circle(fsize)
pen.end_fill()
#endface

#mouth
pen.penup()
pen.goto(x,y_mouth)
pen.down()
pen.fillcolor("#CA9579")
pen.begin_fill()
pen.forward(msize)
pen.right(90)
pen.circle(-msize,180)
pen.right(90)
pen.forward(msize)
pen.end_fill()
#end mouth

#tooth
pen.fillcolor("White")
pen.begin_fill()
pen.forward(tsize)
for i in range(4):
    pen.right(90)
    pen.forward(tsize)
pen.backward(tsize)
for i in range(4):
    pen.right(90)
    pen.forward(tsize)
pen.end_fill()
#end tooth

#eyes
x_reye = x + fsize/6
x_leye = x - fsize/6
y_eye = y + fsize*(29/24)
e = turtle.Turtle()
e.pensize(2)
e.penup()
e.goto(x_reye,y_eye) #right eye begin here
e.pendown()

e.fillcolor("White")
e.begin_fill()
e.right(90)
e.forward(esize)
for i in range(2):
    e.circle(esize,180)
    e.forward(esize)
e.end_fill()

e.penup()
e.goto(x_leye,y_eye) #left eye begin here
e.pendown()
e.fillcolor("White")
e.begin_fill()
#e.right(180)
e.forward(esize)
for i in range(2):
    e.circle(-esize,180)
    e.forward(esize)
e.end_fill()
e.penup()

#eye balls
x_rdot = x + fsize/3
x_ldot = x_rdot - fsize*7/12
dot = esize #eyeball size
pen.penup()
pen.goto(x_rdot,y_eye) #right eye ball here
pen.pendown
pen.dot(dot)
pen.penup()
pen.goto(x_ldot,y_eye) #left eye ball here
pen.pendown
pen.dot(dot)
#end eye balls
turtle.bgcolor("#1FF5FA")
pen.penup()
pen.goto(-300,-300)
e.goto(-300,-300)
#turtle.done()


#turtle.done()

```
