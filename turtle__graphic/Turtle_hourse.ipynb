```python
import turtle
import math
pen = turtle.Turtle()
pen.speed(1)
pen.pensize(3)
pen.penup()
pen.goto(-200,-120) #begin point
pen.pendown()
#frontside of the house
pen.fillcolor("#60EA6E")
pen.begin_fill()
for i in range(2):
    pen.forward(150)
    pen.left(90)
    pen.forward(200)
    pen.left(90)
pen.end_fill()
#end of frontside
#door
pen.forward(90)
pen.left(90)
pen.fillcolor("#E89E3D")
pen.begin_fill()
for i in range(2):
    pen.forward(50)
    pen.left(90)
    pen.forward(30)
    pen.left(90)
pen.end_fill()
#end of the main door
#sideface of the house
pen.right(90)
pen.forward(60)
pen.left(30)
pen.fillcolor("#9A1837")
pen.begin_fill()
ac_ang = 60 #Góc nhọn
obt_ang = 180 - ac_ang #Góc tù
for i in range(3):    
    pen.forward(90)
    pen.left(ac_ang)
    pen.forward(200)
    pen.left(obt_ang)
pen.end_fill()
#roof
#roof
pen.forward(90)
pen.right(30)
pen.fillcolor("#41FAE2")
pen.begin_fill()
for i in range(3):
    pen.forward(150)
    pen.right(120)
pen.end_fill()
#end roof

#roof faceside
pen.right(60)
pen.forward(150)
pen.fillcolor("#E4E30E")
pen.begin_fill()
for i in range(2):
    pen.right(90)
    pen.forward(90)
    pen.right(90)
    pen.forward(150)
pen.end_fill()
#end roof facesize

#window
pen.backward(150)
pen.left(150)
pen.penup()
pen.forward(85)
pen.left(120)
pen.forward(30)
pen.pendown()
pen.fillcolor("pink")
pen.begin_fill()
pen.right(120)
for i in range(2):
    pen.forward(30)
    pen.left(120)
    pen.forward(30)
    pen.left(60)
pen.end_fill()
#end window

###tree
#body
pen.penup()
pen.goto(195,-240)
pen.pendown()
pen.right(180)
tall = 160
wide = tall/10
pen.fillcolor("#81480C")
pen.begin_fill()
for i in range(2):
    pen.forward(tall)
    pen.right(90)
    pen.forward(wide)
    pen.right(90)
pen.end_fill()
###
#leaf
pen.forward(tall)
pen.right(90)
pen.forward(wide/2)
pen.left(90)
pen.fillcolor("#49EB01")
pen.begin_fill()
h = tall/2
for i in range(3):
    pen.penup()
    pen.forward(h)
    pen.pendown()
    pen.left(135)
    pen.forward(h * math.sqrt(2))
    pen.left(135)
    pen.forward(2*h)
    pen.left(135)
    pen.forward(h * math.sqrt(2))
    pen.right(45)
pen.end_fill()
#end tree
#sun
#sun
ssize = 40 #sun size
lr = ssize*2/3
sr = lr*2/3
pen.penup()
pen.goto(180,220)#sun begin here
pen.pendown()
pen.fillcolor("#F95903")
pen.begin_fill()
pen.circle(ssize)
pen.end_fill()
for i in range(4):#sunray
    pen.right(90)
    pen.forward(lr)#long sunray
    pen.backward(lr)
    pen.left(90)
    pen.circle(ssize,45)#draw 1/8 of circle each time
    pen.right(90)
    pen.forward(sr)#short sunray
    pen.backward(sr)
    pen.left(90)
    pen.circle(ssize,45)
#end sun
#turtle.done()
#end of sideface
turtle.done()
```
