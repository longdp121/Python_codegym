```python
import turtle
import math as m

#Khai bao bien
p = turtle.Turtle()
p.speed(1)
p.pensize(3)
x = 0  #Where x start
y = -30  #Where y start
A = 400  #House longside
a = A/2  #House shortside
ci_win = sq_win = A/10  #Circle window size #Squre window size
h_door = A/5  #Main door height
w_door = h_door/2  #Main door wide
roof = a/(m.cos(m.radians(20)))
l_path = h_door/(m.cos(m.radians(10)))  #Path lengh
w_path = (m.sin(m.radians(10))/l_path)*2 + 1.7*w_door  #Path wide
p.penup()
p.goto(x,y)
p.pendown()
#grass
p.penup()
p.left(90)
p.fd(a/2)
p.left(90)
p.fd(A/2 + a/2)
p.fillcolor("#29B204")
p.begin_fill()
for i in range (4):
    p.left(90)
    p.fd(a)
    p.left(90)
    p.fd(A + a)
p.end_fill()
p.goto(x, y)  #back to zero point
p.left(180)  #turn back to right direction
p.pendown()
#end grass

#Main house
p.fd(a)
#Start frontside
p.fillcolor("#E6E7AA")
p.begin_fill()
for i in range (2):  
    p.left(90)
    p.fd(a)
    p.left(90)
    p.fd(A)
p.end_fill()
#end frontside
#Start window
p.fillcolor("#1BA189")
p.begin_fill()
for i in range (2):  #repeat twice to draw pair of windows both side
    p.backward(A/4) #move back for window 
    p.left(90)  #pen heading north
    for i in range (2):  #start a pair of window on one side
        p.penup()
        p.fd(sq_win)
        p.pendown()
        p.fd(sq_win/2)
        for i in range (4):  #repeat 4 time below step for full window
            for i in range (4):  #repeat 4 time for ONE smaller square in each window
                p.fd(sq_win/2)
                p.left(90)
            p.right(90)
        p.fd(sq_win/2)  #go to center top of window after done one
    p.penup()
    p.backward(sq_win*4)  #go back to ground after done upper window
    p.right(90)
    p.backward(A/4)
    p.pendown()
p.end_fill()
#end window
#Start door
p.fd(A/2 - w_door/2)
p.left(90)
p.fillcolor("#EB2828")
p.begin_fill()
for i in range(3):
    p.fd(h_door)
    p.right(90)
    p.fd(w_door)
    p.right(90)
p.end_fill()
p.fd(h_door/2)
p.right(90)
p.penup()
p.fd(w_door/10)
p.pendown()  #door lock start
p.fillcolor("#FFF400")
p.begin_fill()
p.circle(w_door/13)
p.end_fill()
p.penup()  #door lock end
p.fd(w_door*(2/5))
p.right(90)
p.fd(h_door/2)
#end door
#Start circle window
p.fillcolor("#1BA189")
p.begin_fill()
p.fd(sq_win)
p.left(90)
p.pendown()
p.circle(-sq_win/2)
for i in range (2):
    p.circle(-sq_win/2,90)
    p.right(90)
    p.fd(sq_win)
    p.right(90)
p.end_fill()
#end circle window
#roof
p.penup()
p.fd(a)
p.left(90)
p.fd(sq_win)
p.pendown()
p.fillcolor("#E97A14")
p.begin_fill()
p.left(70)
p.fd(roof)
p.left(40)
p.fd(roof)
p.end_fill()
#end roof
#end main house
#path
p.penup()
p.goto(x,y)
p.right(20)
p.pendown()
p.fd(w_door/2)
p.fillcolor("#E48E07")
p.begin_fill()
p.left(80)
p.fd(l_path)
p.left(100)
p.fd(w_path)
p.left(100)
p.fd(l_path)
p.end_fill()
#end path
#Start fence
h_fence = h_door*(2/3)  #heigh of fence
w_fence = a/10  #wide of fence
p.penup()
p.goto(x,y)
p.setheading(0)
p.fd(A/2 + a/2)
p.right(90)
p.fd(a/2)
p.right(180)
p.pendown()
p.left(90)
p.fillcolor("#4C3716")
p.begin_fill()
for i in range (30):
    p.fd(w_fence)
    p.right(90)
    p.fd(h_fence)
    p.right(30)
    p.fd(w_fence)
    p.right(120)
    p.fd(w_fence)
    p.right(30)
    p.fd(h_fence)
    p.right(90)
    p.fd(w_fence)
p.end_fill()
p.left(180)
p.fd(14*w_fence)  #Go to position of fence hinge
p.left(90)
p.fillcolor("#FFFFFF")
p.begin_fill()
for i in range (2):
    p.penup()
    p.fd(h_fence/3)
    p.right(90)
    p.fd(w_fence/4)
    p.pendown()
    for i in range (4):
        p.left(90)
        p.fd(w_fence/2)
    p.backward(w_fence/4)
    p.left(90)
p.end_fill()
p.right(90)
p.penup()
p.fd(2*w_fence)  #Go to position of fence lock
p.pendown()
p.fillcolor("#FFFFFF")
p.begin_fill()
p.circle(w_fence/4)
p.end_fill()
#end fence
#Start sun
ssize = 28 #sun size
lr = ssize*2/3
sr = lr*2/3
p.penup()
p.goto(220,200)#sun begin here
p.pendown()
p.fillcolor("#F95903")
p.begin_fill()
p.circle(ssize)
p.end_fill()
for i in range(4):#sunray
    p.right(90)
    p.forward(lr)#long sunray
    p.backward(lr)
    p.left(90)
    p.circle(ssize,45)#draw 1/8 of circle each time
    p.right(90)
    p.forward(sr)#short sunray
    p.backward(sr)
    p.left(90)
    p.circle(ssize,45)
#end sun
p.penup()
p.goto(-400,-400)
turtle.done()
```
